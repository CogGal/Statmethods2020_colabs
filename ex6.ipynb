There are two independent variables - the Coffee type (Black/Cappucino) and the place of purchase (Social sciences / Forum / Humanities). Each of the 6 groups had n subjects.

Block 7#

```
#@title **<font size="6" color='cornflowerblue'><ins>Question 6.1</ins>** 
#@markdown #**Set the right conditions to get a significant effect for coffee type, without significant effects for the place of purchase or interaction.**

block_id = 7
question_id = 6
required_blocks = range(7, block_id)

def prepare_block():
  try:
    initialized
  except:
    raise Exception("You need to identify yourself in Block 1#")
  verify_dependencies(block_id, required_blocks)


def block7():
  Black_social_sci = -10 #@param {type:"slider", min:-10, max:10}
  Black_forum = -10 #@param {type:"slider", min:-10, max:10}
  Black_humanities = -10 #@param {type:"slider", min:-10, max:10}
  Cappucino_social_sci = -10 #@param {type:"slider", min:-10, max:10}
  Cappucino_forum = -10 #@param {type:"slider", min:-10, max:10}
  Cappucino_humanities = -10 #@param {type:"slider", min:-10, max:10}
  sample_size_n = 10 #@param {type:"slider", min:10, max:150, step:5}
  sd_of_all_groups = 1 #@param {type:"slider", min:1, max:15}

  # Save user answer
  answers[question_id] = (Black_social_sci, Black_forum, Black_humanities, Cappucino_social_sci, Cappucino_forum, Cappucino_humanities, sample_size_n, sd_of_all_groups)

  sample_size=sample_size_n
  means = [Cappucino_social_sci,Cappucino_forum,Cappucino_humanities,Black_social_sci, Black_forum,Black_humanities ]
  sds  = np.repeat(sd_of_all_groups,len(means))

  #Degrees of freedom for interaction: (3-1)(2-1) = 2

  # 1. Plot the 4 groups on a 3D scatter plot (x-axis: homework, y-axis: ice-cream, z-axis:grades). 
  # 2. Plot also the two linear lines of the averages. 

  columns = ["Cappucino","Black"]
  rows = ["Social sciences","Forum","Humanities"]
  ##collect data and binary vectors for defining groups

  coffee = np.concatenate([ np.repeat( columns[i]  ,sample_size*len(rows)) for i in range(len(columns))])
  location = np.concatenate([ np.repeat( rows[i]  ,sample_size*len(columns)/2) for i in range(len(rows))])
  location = np.concatenate([location, location])
  np.random.seed(int(group_number))
  amp = np.concatenate([np.random.normal(means[i], sds[i], sample_size) for i in range(len(means))])

  ##arrange data frame:
  data = {"Coffee":coffee, "Location":location,"Amplitude":amp}
  groupsDF = pd.DataFrame(data, 
                        columns = ["Coffee","Location","Amplitude"])

  ##show intraction term:
  formula = 'Amplitude ~ Coffee + Location + Coffee:Location'
  model = sm.formula.ols(formula, data=groupsDF).fit()

  # Perform ANOVA and print table
  aov_table = sm.stats.anova_lm(model, typ=2)
  alpha = 0.05 #Or whatever you want your alpha to be.
  df1_1 = aov_table.df[2]
  df2_1 = aov_table.df[3]
  Fint_1 = aov_table.F[2]
  a_v = aov_table["PR(>F)"]

  del aov_table["PR(>F)"]

  sampled_Tmean = np.mean(data["Amplitude"])
  sampled_meansRows = [np.mean(data["Amplitude"][data["Location"]=="Social sciences"]),
                      np.mean(data["Amplitude"][data["Location"]=="Forum"]),
                      np.mean(data["Amplitude"][data["Location"]=="Humanities"])]
  sampled_meansCols = [np.mean(data["Amplitude"][data["Coffee"]=="Black"]),
                      np.mean(data["Amplitude"][data["Coffee"]=="Cappucino"])]
  sampled_means = [np.mean(data["Amplitude"][(data["Location"]=="Social sciences")  & (data["Coffee"]=="Black")]),
                  np.mean(data["Amplitude"][(data["Location"]=="Forum")  & (data["Coffee"]=="Black")]),
                  np.mean(data["Amplitude"][(data["Location"]=="Humanities")  & (data["Coffee"]=="Black")]),
                  np.mean(data["Amplitude"][(data["Location"]=="Social sciences") & (data["Coffee"]=="Cappucino")]),
                  np.mean(data["Amplitude"][(data["Location"]=="Forum")  & (data["Coffee"]=="Cappucino")]),
                  np.mean(data["Amplitude"][(data["Location"]=="Humanities") & (data["Coffee"]=="Cappucino")])]

  print("Mean of all groups together (X) = ", round(sampled_Tmean,3))
  print("")
  from tabulate import tabulate
  print(tabulate([['Social sciences (X_1.)', sampled_means[0],sampled_means[3],sampled_meansRows[0]],
                  ['Forum (X_2.)', sampled_means[1],sampled_means[4],sampled_meansRows[1]],
                  ['Humanities',sampled_means[2],sampled_means[5],sampled_meansRows[2]],
                  ['Mean', sampled_meansCols[0],sampled_meansCols[1],sampled_Tmean] ],
                headers=['Location\Coffee', 'Black ( X_.1 )', "Cappucino ( X_.2 )", "Mean"]))
  new_index=list(aov_table.index)
  new_index[3] = "Within groups"
  aov_table.index = new_index

  print("\n\nTwo way anova results:")
  print(np.round(aov_table,3),"\n\n")

  g = sns.catplot(x="Location", y="Amplitude", hue="Coffee",
                  errwidth = 2, capsize=.1, palette=["sandybrown","Black"], height=9, aspect=1.5,
                  kind="point", data=groupsDF)
  g.despine(left=True);
  plt.ylabel ("Amplitude(µV)")
  plt.title("2 by 3 design: Coffee type (Black/Cappucino) and location of purchase (Social sciences / Forum / Humanities)");
  g.set_xticklabels(["Social sciences","Forum","Humanities"]);
  for i in range(3):
    interactive_answers[i] = a_v[i]
  return a_v


try:
  prepare_block()
  # Get GLOBAL answer from user
  # Verify GLOBAL answer is valid
  # Process
  # Save GLOBAL answer
  # Message to user
  a_v = block7() # Use return values, if any, to create global variables
  # Mark block as run
  blocks_run[block_id] = True
except Exception as e:
  print(str(e))
```
